```python
#importing dependencies
import numpy as np
```

```python
class LogisticRegression():
  #initiating the parameters (learning rate and number of iterations)
  def __init__(self, alpha, epochs):
    self.alpha = alpha
    self.epochs = epochs
  def fit(self, X, Y):
    #number of training examples and number of features
    self.m, self.n = X.shape

    #initiating the weights and bias
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    #implementing gradient descent
    for i in range(self.epochs):
      self.update_weights()

  def update_weights(self):
    Y_prediction = 1 / (1 + np.exp(-np.dot(self.X, self.w) + self.b))

    #calculate the gradients

    dw = (1 / self.m) * np.dot(self.X.T, (Y_prediction - self.Y))
    db = (1 / self.m) * np.sum(Y_prediction - self.Y)

    #update the weights
    self.w = self.w - self.alpha * dw
    self.b = self.b - self.alpha * db

  def predict(self, X):
    Y_pred = (1 / (1 + np.exp(-np.dot(X, self.w) + self.b)))
    Y_pred = np.where(Y_pred > 0.5, 1, 0)
    return Y_pred
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

###Data Collection and Analysis

```python
diabetes_df = pd.read_csv('/content/diabetes (1).csv')
```

```python
diabetes_df.head()
```

```python
diabetes_df.shape
```

```python
diabetes_df.isnull().sum()
```

```python
diabetes_df['Outcome'].value_counts()
```

```python
positive = diabetes_df[diabetes_df['Outcome'] == 1]
negative = diabetes_df[diabetes_df['Outcome'] == 0]
```

```python
negative_sample = negative.sample(n=268)
```

```python
diabetes_df = pd.concat([positive, negative_sample])
diabetes_df['Outcome'].value_counts()
```

```python
diabetes_df.describe()
```

```python
features = diabetes_df.drop(columns='Outcome', axis=1)
target = diabetes_df['Outcome']
```

```python
print(features)
```

```python
print(target)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, X_test.shape)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
X_train.std()
```

```python
X_test.std()
```

###Training the model


```python
model = LogisticRegression(alpha=0.01, epochs=1000)
model.fit(X_train, Y_train)
```

```python
training_prediction = model.predict(X_train)
training_accuracy = accuracy_score(training_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(test_prediction, Y_test)
```

```python
print('Accuracy on test data: ', test_accuracy)
```
